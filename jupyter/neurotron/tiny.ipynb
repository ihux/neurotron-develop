```python
from neurotron import Cells, Text, Trainer, Toy

toy = Toy('Tiny')
toy.text = Text(toy.raw,8);  print(toy.text)
cells = Cells((8,25,150,4),3,char=True)
train = Trainer(cells)
N = 100
for k in range(20): print(toy.text(k))
for k in range(N): train(toy.text(k))
```
