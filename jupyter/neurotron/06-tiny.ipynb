```python
from neurotron import Cells, Text, Trainer, Toy

toy = Toy('Tiny')
toy.text = Text(toy.raw,11);  print(toy.text)
cells = Cells((6,12,20,4),3,char=True)
train = Trainer(cells)
N = 3
for k in range(N): print(toy.text(k))
for k in range(N): train(toy.text(k))
train.show()
```
